##Adding the files, starting with athletes.

```python
file_path = "/content/drive/MyDrive/Courses/CIS531/Term_Project/olympics/athletes.csv"
data = pd.read_csv(file_path)
```

```python
data.head()
```

Choosing relevant columns (didn't leave much out)

```python
athletes_columns = [
    'code','name', 'name_tv', 'gender', 'function', 'country_code','country_long', 'nationality', 'height', 'weight',
    'disciplines', 'events', 'birth_date', 'birth_place', 'birth_country', 'residence_place',
    'residence_country', 'nickname', 'hobbies', 'education', 'family', 'lang', 'coach', 'reason',
    'hero', 'influence', 'philosophy', 'sporting_relatives', 'ritual', 'other_sports'
]
```

```python
athletes_data = data[athletes_columns]
```

Filling missing values with empty strings

```python
athletes_data = athletes_data.fillna("")
```

Normalization: Converting all text to lowercase

```python
athletes_data = athletes_data.map(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
athletes_data.head()
```

Add labels for each field and turn it into a single string so I can train on it better


```python
athletes_data['text'] = athletes_data.apply(
    lambda row: ' | '.join([f"{col}: {row[col]}" for col in row.index]), axis=1
)
```

```python
athletes_data = athletes_data[['text']]
```

```python
athletes_data.head(150)
```

Save the preprocessed set to a new file

```python
athletes_data.to_csv('/content/drive/MyDrive/Courses/CIS531/Term_Project/olympics/Processed_Data/athletes_processed.csv', index=False)
```
